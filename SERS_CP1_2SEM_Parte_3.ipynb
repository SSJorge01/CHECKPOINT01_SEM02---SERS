27. Distribuição do consumo
- Crie histogramas e séries temporais para a variável Appliances.
- Pergunta: o consumo tende a se concentrar em valores baixos ou altos?

```python
df['Appliances'].hist(bins=50)
plt.title('Distribuição do Consumo')
plt.xlabel('Consumo')
plt.ylabel('Frequência')

''' O consumo tende a se concentrar em valores mais baixos, sendo que a maior
parte está entre o consumo '0' e '200'. */ '''
```

28. Correlações com variáveis ambientais
- Calcule correlações entre Appliances e variáveis como temperatura e umidade.
- Pergunta: quais fatores têm mais relação com o consumo?

```python
correlation_matrix = df[['Appliances', 'T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'RH_out', 'Tdewpoint']].corr()

display(correlation_matrix['Appliances'].sort_values(ascending=False))

''' Os fatores que mais tem relação com o consumo seriam o T2(Positivo) e o
RH_out (Negativo).'''
```

29. Normalização dos dados
- Aplique Min-Max Scaling às variáveis numéricas.
- Reutilize esses dados em modelos posteriores.

```python
from sklearn.preprocessing import MinMaxScaler

# Selecionar apenas as colunas numéricas, excluindo a coluna 'date'
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Inicializar o MinMaxScaler
scaler = MinMaxScaler()

# Aplicar o Min-Max Scaling nas colunas numéricas
df1 = df.copy()
df1[numeric_cols] = scaler.fit_transform(df1[numeric_cols])

# Exibir as primeiras linhas do DataFrame normalizado
display(df1.head())
```
